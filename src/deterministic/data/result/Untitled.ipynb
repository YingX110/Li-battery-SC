```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
```

```python
data = pd.read_csv('Scen_Efficiency00.csv', index_col=0)
eff = data.iloc[:, 0:16]
capacity = data.iloc[:, [16]]
```

```python
sns.set_theme(rc={'figure.figsize':(11,8)})
sns.set_style("white")
line_eff = sns.lineplot(data=eff.T)

fig = line_eff.get_figure()
fig.savefig("./plots/Li_Eff.svg") 
```
