```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
```

```python
ecoval = pd.read_csv('eco_val.csv')
ecocount = pd.read_csv('eco_count.csv')
ecocount
```

```python
ecoval.at[4, 'lon1']
```

```python
def plotnetwork(data, name):
    m = Basemap()
    m.drawcoastlines()
    m.drawcountries()

    for i in range(data.shape[0]):
        lon1 = data.at[i, 'lon1']
        lat1 = data.at[i, 'lat1']
        lon2 = data.at[i, 'lon2']
        lat2 = data.at[i, 'lat2']
        if (lon1, lat1) != (lon2, lat2):
            m.drawgreatcircle(lon1, lat1, lon2, lat2, linewidth=2, color='r')
#             m.drawgreatcircle(lon1, lat1, lon2, lat2, linewidth=2, color='r')
        else:
            pass
    plt.savefig(name)
```

```python
plotnetwork(ecoval, 'eco_val_nw.svg')
```

```python
plotnetwork(ecocount, 'eco_count_nw.svg')
```

```python
m = Basemap(projection='robin',lon_0=0,resolution='c')
m.drawcoastlines()
m.fillcontinents(color='coral',lake_color='aqua')
# draw parallels and meridians.
m.drawparallels(np.arange(-90.,120.,30.))
m.drawmeridians(np.arange(0.,360.,60.))
m.drawmapboundary(fill_color='aqua')
plt.title("Robinson Projection")
plt.savefig("test2.svg")
```
