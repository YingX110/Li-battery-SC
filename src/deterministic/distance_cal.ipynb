```python
import geopy.distance
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('./data/capacity2.csv')
country_name = df['country'].tolist()
```

```python
country_name
```

```python
ncty = len(country_name)
res = np.zeros([ncty,ncty])
for i in range(ncty):
    for j in range(ncty):
        c1 = country_name[i]
        c2 = country_name[j]
        c1info = geopy.geocoders.Nominatim(user_agent="my-app").geocode(c1)[1]
        c2info = geopy.geocoders.Nominatim(user_agent="my-app").geocode(c2)[1]
        distance = geopy.distance.distance(c1info, c2info).km
        res[i,j] = distance
        
```

```python
df = pd.DataFrame(res, columns=country_name)
df.index = country_name
df.to_csv('distance_31111.csv')
```
